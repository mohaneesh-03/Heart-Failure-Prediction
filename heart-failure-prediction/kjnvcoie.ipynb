```python
import pandas as pd
from tensorflow import keras
import matplotlib as mp
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.replace({'M':1, 'F':0},inplace=True)
```

```python
df.replace({'Y':1, 'N':0},inplace=True)
```

```python
df.replace({'ATA':-2, 'NAP':-1, 'ASY':1, 'TA':2},inplace=True)
df.replace({'Up':-1, 'Down':1, 'Flat':0},inplace=True)
df.replace({'Normal':-1, 'ST':0, 'LVH':1},inplace=True)
```

```python
df
```

```python
for i in df:
    print(df[i].unique())
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
df.head
```

```python
df.head(7)
```

```python
col_scale=['Age', 'ChestPainType', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
df[col_scale]=scalar.fit_transform(df[col_scale])
```

```python
df.head(7)
```

```python
x = df.drop('HeartDisease', axis='columns')
y=df['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
len(X_train.columns)
```

```python
X_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(12 , input_shape=(11,) , activation='relu'),
    keras.layers.Dense(8,activation='relu'),
    keras.layers.Dense(6,activation='relu'),
    keras.layers.Dense(4,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid'),
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train , y_train , epochs=1000)
```

```python
y_pred  =model.predict(X_test)
y_pred[:10]
```

```python
y[:10]
```

```python
model.evaluate(X_test,y_test)
```
